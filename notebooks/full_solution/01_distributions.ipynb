# AirBnB data EDA: feature distributions

## Notebook set-up

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 1. Data loading

Load the AirBnB NYC 2019 dataset from the remote CSV file and examine its basic structure. We'll display the first few rows transposed to get a better view of all columns and their data types.

```python
data_df = pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/data-preprocessing-project-tutorial/main/AB_NYC_2019.csv')
data_df.head().transpose()
```

```python
data_df.info()
```

The dataset contains 48,895 AirBnB listings with 16 features including location data (latitude, longitude, neighbourhood), listing details (price, room_type, minimum_nights), and review information. We can see there are some missing values, particularly in the 'reviews_per_month' column.

## 2. Categorical features

### 2.1. Levels

Analyze categorical features to understand the diversity and distribution of categories. We'll examine how many unique levels each categorical feature has, which helps us understand the complexity and potential encoding requirements for each feature.

```python
categorical_features = ['neighbourhood_group','neighbourhood','room_type','host_name']

print('Unique levels by feature')
data_df[categorical_features].nunique()
```

The categorical features show varying complexity - neighbourhood_group has only 5 levels (manageable), room_type has 3 levels (simple), while neighbourhood has 221 unique areas and host_name has 11,452 unique hosts. The high cardinality of neighbourhood and host_name suggests these may need special handling in modeling.

### 2.2. Level counts

Visualize the distribution of categorical features to understand class imbalances and data patterns. For features with many categories (neighbourhood, host_name), we'll use area plots with log scale to handle the wide range of frequencies.

```python
fig, axs = plt.subplots(2,2, figsize=(6,6))
axs = axs.flatten()

fig.suptitle('Listing attribute level counts')

# Plot neighborhood group level counts
level_counts = data_df['neighbourhood_group'].value_counts()

axs[0].set_title('Neighborhood group')
axs[0].bar(list(range(len(level_counts))), level_counts, tick_label=level_counts.index, color='black')
axs[0].tick_params(axis='x', labelrotation=45)
axs[0].set_ylabel('Listings')

# Plot room type level counts
level_counts = data_df['room_type'].value_counts()

axs[1].set_title('Room type')
axs[1].bar(list(range(len(level_counts))), level_counts, tick_label=level_counts.index, color='black')
axs[1].tick_params(axis='x', labelrotation=45)
axs[1].set_ylabel('Listings')

# Plot neighborhood level counts
level_counts = data_df['neighbourhood'].value_counts()

axs[2].set_title('Neighborhood')
axs[2].fill_between(list(range(len(level_counts))), level_counts, color='black')
axs[2].set_xlabel('Neighborhood number')
axs[2].set_ylabel('Listings')
axs[2].set_yscale('log')

# Plot neighborhood level counts
level_counts = data_df['host_name'].value_counts()

axs[3].set_title('Host name')
axs[3].fill_between(list(range(len(level_counts))), level_counts, color='black')
axs[3].set_xlabel('Host number')
axs[3].set_ylabel('Listings')
axs[3].set_yscale('log')

plt.tight_layout()
```

The visualizations reveal important patterns:
- **Neighbourhood groups**: Manhattan and Brooklyn dominate with ~19K and ~20K listings respectively
- **Room types**: "Entire home/apt" is most common (~25K), followed by "Private room" (~22K)
- **Neighbourhoods**: Highly skewed distribution with a few popular areas having many listings
- **Hosts**: Most hosts have only 1-2 listings, but some "super hosts" manage many properties

## 3. Numerical features

### 3.1. Descriptive statistics

Examine the descriptive statistics of numerical features to understand their central tendencies, spread, and identify potential outliers. This analysis will help us understand data quality issues and inform preprocessing decisions.

```python
numeric_features = ['latitude','longitude','price','minimum_nights','number_of_reviews','last_review','reviews_per_month','calculated_host_listings_count','availability_365']

data_df[numeric_features].describe().transpose()
```

The statistics reveal several important insights:
- **Price**: Highly variable (mean $152, max $10,000) with likely extreme outliers
- **Minimum nights**: Most require 1-3 nights, but max is 1,250 nights (possible data errors)
- **Reviews**: Many listings have 0 reviews, suggesting new or unpopular listings
- **Reviews per month**: Only ~38K listings have data (10K missing), indicating inactive listings
- **Host listings**: Most hosts have 1-2 properties, but one has 327 listings
- **Availability**: Wide range suggests different listing strategies (full-time vs occasional)

### 3.2. Distribution plots

#### 3.2.1. Latitude & longitude

Analyze the spatial distribution of AirBnB listings by examining latitude and longitude coordinates. We'll create histograms for each coordinate and a scatter plot to visualize the geographic spread of listings across NYC's five boroughs.

```python
fig, axs = plt.subplots(1,3, figsize=(12,4))
axs = axs.flatten()

fig.suptitle('Listing location distributions')

axs[0].set_title('Latitude')
axs[0].hist(data_df['latitude'], color='black', bins=20)
axs[0].set_xlabel('Degrees')
axs[0].set_ylabel('Listings')

axs[1].set_title('Longitude')
axs[1].hist(data_df['longitude'], color='black', bins=20)
axs[1].set_xlabel('Degrees')
axs[1].set_ylabel('Listings')

axs[2].set_title('Location')
for neighborhood_group in data_df['neighbourhood_group'].unique():
    axs[2].scatter(
        data_df['longitude'][data_df['neighbourhood_group'] == neighborhood_group],
        data_df['latitude'][data_df['neighbourhood_group'] == neighborhood_group],
        s=0.2,
        label=neighborhood_group
    )
axs[2].set_xlabel('Longitude')
axs[2].set_ylabel('Latitude')
axs[2].legend(loc='upper left', markerscale=20)

plt.tight_layout()
plt.show()
```

The location analysis shows clear geographic clustering:
- **Latitude distribution**: Multi-modal with peaks corresponding to different boroughs
- **Longitude distribution**: Relatively normal distribution centered around NYC's longitude
- **Spatial scatter**: Clearly shows NYC's geography with Manhattan as a dense vertical cluster, Brooklyn/Queens as horizontal spread, and Staten Island as a separate cluster. The color coding by neighbourhood group makes the five boroughs visually distinct.

#### 3.2.2. Other listing attributes

Visualize the distributions of key listing attributes (price, minimum nights, reviews, etc.) using histograms with log scale. This will help us identify the shape of distributions and spot potential outliers or unusual patterns that may require data cleaning.

```python
features = ['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']
plot_titles = ['Price', 'Minimum booking', 'Total reviews', 'Review rate', 'Listings per host', 'Vacancies last year']
xlabels = ['USD', 'Nights', 'Reviews', 'Reviews per month', 'Listings per host', 'Available nights']

fig, axs = plt.subplots(2, 3, figsize=(10, 6))
axs = axs.flatten()

fig.suptitle('Listing attribute distributions')

for i, (plot_title, feature, xlabel) in enumerate(zip(plot_titles, features, xlabels)):

    axs[i].set_title(plot_title)
    axs[i].hist(data_df[feature], bins=20, color='black')
    axs[i].set_xlabel(xlabel)
    axs[i].set_ylabel('Listings')
    axs[i].set_yscale('log')

plt.tight_layout()
```

The distribution plots reveal several patterns and potential data quality issues:
- **Price**: Right-skewed with most listings under $200, but extreme outliers up to $10K
- **Minimum nights**: Heavily right-skewed with most requiring 1-7 nights, but suspicious values like 999+ nights
- **Total reviews**: Many zero-review listings, suggesting new/inactive properties
- **Review rate**: Missing data for ~10K listings, bimodal distribution for active listings
- **Listings per host**: Highly right-skewed - most hosts have 1-2 listings, few have 50+
- **Availability**: Bimodal distribution with peaks at 0 (unavailable) and 365 (always available)
